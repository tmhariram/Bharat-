```python
def plot_histplots_subplots(df):
    num_rows = 6
    num_cols = 2

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 30))

    for i, column in enumerate(df.columns[:-1]):
        row = i // num_cols
        col = i % num_cols
        ax = axes[row, col]

        sns.histplot(df[column], ax=ax)
        ax.set_title(f"Histogram of {column}")

    plt.tight_layout()
    plt.show()

plot_histplots_subplots(df)
```

```python
X=df.drop('quality',axis=1)
y=df['quality']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error,mean_squared_error,DistanceMetric

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler=StandardScaler()
```

```python
scaled_X_train=scaler.fit_transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

```python
lr_model=LinearRegression()
knn_model=KNeighborsRegressor()
svr_model=SVR()
rfr_model=RandomForestRegressor()
```

```python
lr_model.fit(X_train,y_train)
knn_model.fit(X_train,y_train)
svr_model.fit(scaled_X_train,y_train)
rfr_model.fit(X_train,y_train)
```

```python
lr_grid={'fit_intercept': [True,False],
         'copy_X':[True,False],
         'n_jobs':[2,4,6],
         'positive':[True,False]}

knn_grid={'weights' : ['uniform', 'distance'],
          'n_neighbors':[2,5,10,50,100],
          'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
          'leaf_size':[2,3,4,5,7,8,10],
          'p':[3,5,7,9,10,50,100]}

svr_grid={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
          'degree':[1,2,4],
          'gamma':['auto','scale'],
          'C':[0.1,0.2,0.5],
          'max_iter':[1,2,20,100]}

rfr_grid={'n_estimators':[1,10,50,100,150,200],
          'criterion':["squared_error", "absolute_error", "poisson"],
          'max_features':["auto", "sqrt", "log2"]}
```

```python
lr_model1=GridSearchCV(lr_model,param_grid=lr_grid,cv=2,verbose=2)
knn_model1=GridSearchCV(knn_model,param_grid=knn_grid,cv=2,verbose=2)
svr_model1=GridSearchCV(svr_model,param_grid=svr_grid,cv=2,verbose=2)
rfr_model1=GridSearchCV(rfr_model,param_grid=rfr_grid,cv=2,verbose=2)
```

```python
lr_model1.fit(X_train,y_train)
knn_model1.fit(X_train,y_train)
svr_model1.fit(scaled_X_train,y_train)
rfr_model1.fit(X_train,y_train)
```

```python
print('\033[1m'+'Best parameters for Linear Regression model are'+'\033[0m \n',lr_model1.best_params_)
print('\033[1m'+'Best parameters for K Nearest Neighbors model are'+'\033[0m \n',knn_model1.best_params_)
print('\033[1m'+'Best parameters for Support Vector model are'+'\033[0m \n',svr_model1.best_params_)
print('\033[1m'+'Best parameters for Random Forest model are'+'\033[0m \n',rfr_model1.best_params_)

```

```python
final_lr_model=LinearRegression(copy_X=True,fit_intercept=False,n_jobs=2,positive=False)
final_knn_model=KNeighborsRegressor(algorithm='auto',leaf_size=2,n_neighbors=10,p=3,weights='distance')
final_svr_model=SVR(C=0.5,degree=1,gamma='auto',kernel='rbf',max_iter=100)
final_rfr_model=RandomForestRegressor(criterion='squared_error',max_features='sqrt',n_estimators=200)
```

```python
final_lr_model.fit(X_train,y_train)
final_knn_model.fit(X_train,y_train)
final_svr_model.fit(X_train,y_train)
final_rfr_model.fit(X_train,y_train)
```

```python
MAE=[]
RMSE=[]
```

```python
model_list=[lr_model,knn_model,svr_model,rfr_model]
MAE=[]
RMSE=[]

for model in model_list:
    preds=model.predict(X_test)
    mae=mean_absolute_error(y_test,preds)
    rmse=np.sqrt(mean_squared_error(y_test,preds))
    MAE.append(mae)
    RMSE.append(rmse)
```

```python
final_model_list=[lr_model1,knn_model1,svr_model1,rfr_model1]
MAE1=[]
RMSE1=[]

for model in final_model_list:
    preds=model.predict(X_test)
    mae=mean_absolute_error(y_test,preds)
    rmse=np.sqrt(mean_squared_error(y_test,preds))
    MAE1.append(mae)
    RMSE1.append(rmse)
```

```python
model_list=['lr_model','knn_model','svr_model','rfr_model']

pd.DataFrame({'models':model_list,
              'MAE':MAE,
              'RMSE':RMSE})
```
